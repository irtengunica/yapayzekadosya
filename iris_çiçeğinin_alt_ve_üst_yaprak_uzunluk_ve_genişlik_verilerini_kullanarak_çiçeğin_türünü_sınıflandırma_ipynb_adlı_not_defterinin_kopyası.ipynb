bağımlı ve bağımsız değişkenleri oluşturmak
için bağımsız nitelikleri X, bağımlı niteliği Y değişkenine atıyoruz.

```python
X = iris.data
Y = iris.target
```

 Eğitim ve test verileri ayırma (veri seti genellikle %75’den başlayarak %80, %85’e
kadar ayrılırken test veri seti ise %15 ile %25 arasında değişir. Modelde eğitim ve test verileri
yüzdeleri kullanılan veri setine bağlı olarak değişebilmektedir.)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 0)
print("Eğitim veri seti boyutu=",len(X_train))
print("Test veri seti boyutu=",len(X_test))

```

scikit-learn kütüphanesinden karar ağaçları sınıflandırıcısını çağırır ve
model adında bir değişkene aktarır.Ağaç yapısında dallardaki birleşme noktaları (düğümler) bir özelliğe (özniteliğe) tekabül etmekte,
her bir dal ise bir kararı işaret etmektedir. Karar ağacında düğümler, sınıflandırılacak gruptaki
özellikleri temsil eder ve dallar ise düğümlerin alabileceği değerleri temsil eder.


```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

Python programlama editöründe iris çiçeği türlerine ait
eğitim verilerinin model.fit komutu ile eğitimini gerçekleştirir. Fit
eğrisi üzerine verileri mümkün olan en yakın noktaya yerleştirmek için model eğitilir.
image.png

```python
model.fit(X_train,Y_train)
Y_tahmin = model.predict(X_test)
```

 karar ağaçları modeli ile eğitilen iris çiçeği sınıfını
hata matrisi (confusion matrix) kullanarak modelin başarısını ölçmektir. Hata matrisi yapay zekâ
sınıflandırmaları için performans ölçer.

```python
from sklearn.metrics import confusion_matrix
hata_matrisi = confusion_matrix(Y_test, Y_tahmin)
print(hata_matrisi)
```

Iris Çiçeği Sınıflandırma - Farklı Yöntemler
1. Lojistik Regresyon

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Iris veri setini yükleme
iris = load_iris()
X = iris.data
Y = iris.target

# Veriyi eğitim ve test kümelerine ayırma
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=0)

# Lojistik Regresyon modelini oluşturma ve eğitme
model = LogisticRegression()
model.fit(X_train, Y_train)

# Test verisi üzerinde tahmin yapma
Y_tahmin = model.predict(X_test)

# Doğruluk skorunu hesaplama
dogruluk = accuracy_score(Y_test, Y_tahmin)
print("Doğruluk:", dogruluk)
from sklearn.metrics import confusion_matrix
hata_matrisi = confusion_matrix(Y_test, Y_tahmin)
print(hata_matrisi)
```

2. Destek Vektör Makineleri (SVM):

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Iris veri setini yükleme
iris = load_iris()
X = iris.data
Y = iris.target

# Veriyi eğitim ve test kümelerine ayırma
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=0)

# SVM modelini oluşturma ve eğitme
model = SVC()
model.fit(X_train, Y_train)

# Test verisi üzerinde tahmin yapma
Y_tahmin = model.predict(X_test)

# Doğruluk skorunu hesaplama
dogruluk = accuracy_score(Y_test, Y_tahmin)
print("Doğruluk:", dogruluk)
from sklearn.metrics import confusion_matrix
hata_matrisi = confusion_matrix(Y_test, Y_tahmin)
print(hata_matrisi)
```

3. Rastgele Orman:

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Iris veri setini yükleme
iris = load_iris()
X = iris.data
Y = iris.target

# Veriyi eğitim ve test kümelerine ayırma
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=0)

# Rastgele Orman modelini oluşturma ve eğitme
model = RandomForestClassifier()
model.fit(X_train, Y_train)

# Test verisi üzerinde tahmin yapma
Y_tahmin = model.predict(X_test)

# Doğruluk skorunu hesaplama
dogruluk = accuracy_score(Y_test, Y_tahmin)
print("Doğruluk:", dogruluk)
from sklearn.metrics import confusion_matrix
hata_matrisi = confusion_matrix(Y_test, Y_tahmin)
print(hata_matrisi)
```

5. Yapay Sinir Ağı ile Iris Çiçekleri Sınıflandırma

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

# Iris veri setini yükleme
iris = load_iris()
X = iris.data
Y = iris.target

# Verileri standartlaştırma (normalleştirme)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Sınıf etiketlerini one-hot encoding'e dönüştürme
Y_categorical = to_categorical(Y)

# Veriyi eğitim ve test kümelerine ayırma
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y_categorical, test_size=0.20, random_state=0)

# Yapay Sinir Ağı modelini oluşturma
model = Sequential()
model.add(Dense(10, activation="relu", input_shape=(X_train.shape[1],)))  # 10 gizli düğüm, relu aktivasyon
model.add(Dense(3, activation="softmax"))  # 3 çıkış düğümü (sınıf sayısı), softmax aktivasyon

# Model derleme
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

# Model eğitimi
model.fit(X_train, Y_train, epochs=200, batch_size=32)

# Test verisi üzerinde değerlendirme
loss, accuracy = model.evaluate(X_test, Y_test)
print("Test doğruluk:", accuracy)
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import scale
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Digits datasetini yükleme
digits = load_digits()

# X (özellikler) ve y (sınıf etiketleri) veri kümelerini ayırma
X = digits.data
y = digits.target

# Görüntüleri görselleştirme (isteğe bağlı)
for i in range(10):
  plt.subplot(5, 5, i + 1)
  plt.imshow(X[i].reshape(8, 8), cmap="gray")
  plt.title(y[i])
  plt.axis("off")
plt.show()

# Görsel ön işleme (normalize etme)
X_scaled = scale(X)

# Veriyi eğitim ve test kümelerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.20, random_state=42)

# KNN sınıflandırıcı oluşturma ve eğitme
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Test veri kümesi üzerinde tahmin yapma
y_pred = knn.predict(X_test)

# Doğruluk skorunu hesaplama
accuracy = accuracy_score(y_test, y_pred)
print("Doğruluk:", accuracy)
```
